```python
from cpymad.madx import Madx
from cpymad.madx import Sequence

%config InlineBackend.figure_format = 'retina'
```

# MAD-X environment setup

```python
madx = Madx()
madx.command.beam('PARTICLE=PROTON, PC = 2.14;')
madx.input('BRHO := BEAM->PC * 3.3356;')
# call sequence of main units
madx.call('/eos/home-a/ahuschau/www/test-acc-models/repository/PS/2019/PS_MU.seq')
# call sequence of straight section elements
madx.call('/eos/home-a/ahuschau/www/test-acc-models/repository/PS/2019/PS_SS.seq')
# call general strength file
madx.call('/eos/home-a/ahuschau/www/test-acc-models/repository/PS/2019/PS.str')
madx.call('/eos/home-a/ahuschau/www/test-acc-models/repository/PS/2019/scenarios/Bare_machine/0_proton_injection_energy/PS_PRO_bare_machine.str')
```

# Expansion of the sequence

```python
seq = Sequence('ps', madx)
seq.expand()

element_classes = []

for elem in seq.expanded_elements:
    if elem.name[2] == '.':
        element_classes = element_classes + [elem.parent.name]

element_classes = list(set(element_classes))
```

# Print element classes to be removed to file

```python
filename = 'remove_elements.seq'
    
# elements to be kept in the sequence
excluded = ['MQNAAIAP'.lower(), # normal LEQ
            'MQNABIAP'.lower(), # wide LEQ
            'ACC10'.lower(),
            'BWSRB'.lower(), # wire scanners
            'sbend']

with open(filename, 'w') as f:
    f.write("select, flag=seqedit, clear;\n")
    f.write("select, flag=seqedit, pattern = '^PA\.C10\.[1-8]';      ! remove all but one 10 MHz cavities\n")
    f.write("select, flag=seqedit, pattern = '^PA\.C10\.96';         ! remove all but one 10 MHz cavities\n")

    for c in element_classes:
        if c not in excluded:
            f.write('select, flag=seqedit, class = ' + c.upper() + ';\n')
```
